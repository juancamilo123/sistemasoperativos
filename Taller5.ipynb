```python
import numpy as np
import matplotlib.pyplot as plt

cont=1
a=pow(5, 5)
c=12345
m=pow(2, 31)
xo=25
xo2=100
generador_ui=[]
generador_ui2=[]


def aplicar_formula(generador_ui,generador_ui2):
    x=[]
    y=[]
    
    for x1,y1 in zip(generador_ui,generador_ui2):
        x.append(2*x1-1)
        y.append(2*y1-1)
        
    return x,y

def puntos(x,y):
    puntos_adent_x=[]
    puntos_adent_y=[]
    puntos_afuera_x=[]
    puntos_afuera_y=[]
    puntos_bordes_x=[]
    puntos_bordes_y=[]
    cant_puntos_adentros=0
    for x1,y1 in zip(x,y):
            if (pow(x1,2)+pow(y1,2))<1:
                puntos_adent_x.append(x1)
                puntos_adent_y.append(y1)
                cant_puntos_adentros=cant_puntos_adentros+1
            elif (pow(x1,2)+pow(y1,2))>1:
                puntos_afuera_x.append(x1)
                puntos_afuera_y.append(y1)
            elif (pow(x1,2)+pow(y1,2))==1:
                puntos_bordes_x.append(x1)
                puntos_bordes_y.append(y1)
                
    return (puntos_adent_x,puntos_adent_y,puntos_afuera_x,
            puntos_afuera_y,puntos_bordes_x,
            puntos_bordes_y,cant_puntos_adentros)

#Congruencia mixta
def congru_mixto(a,c,m,xo):       
    xn=(5*xo+c)%m
    un=xn/m
    return xn,un 


print ("Ingrese la cantidad de valores:")
cant_x=int(input())

#Obtiene un array de números generados 
while cont<=cant_x:  
    xo,un=congru_mixto(a,c,m,xo)
    generador_ui.append(un)
    
    xo2,un=congru_mixto(a,c,m,xo2)
    generador_ui2.append(un)
    cont=cont+1
    
x,y=aplicar_formula(generador_ui,generador_ui2)

puntos_adent_x,puntos_adent_y,puntos_afuera_x,puntos_afuera_y,puntos_bordes_x,puntos_bordes_y,cant_puntos_adentros=puntos(x,y)

plt.scatter(puntos_adent_x,puntos_adent_y,color='r')
plt.scatter(puntos_afuera_x,puntos_afuera_y,color='b')
plt.scatter(puntos_bordes_x,puntos_bordes_y,color='g')
plt.show()

```
